```python
import mysql.connector
import pandas as pd
from tabulate import tabulate 
import matplotlib.pyplot as plt

# fonction pour se connecter Ã  la base de donnÃ©es 
def connect_to_database():
    conn = mysql.connector.connect(                                                     
        host="localhost",
        port=3306,
        user="root",
        password="upsidedown",
        database="formation_numcity"
    )
    print("\nTape 'Stop' pour fermer la connexion.\n")

    while True:
        commande = input("Entrez la commande SQL : ")

        if commande.strip().lower() == "stop":
            print("âœ… Connexion fermÃ©e.")
            conn.close()
            break

        try:
            cursor = conn.cursor()
            cursor.execute(commande)

            # RÃ©cupÃ©ration et affichage organisÃ© avec pandas
            rows = cursor.fetchall()
            columns = [desc[0] for desc in cursor.description]
            df = pd.DataFrame(rows, columns=columns)

            if df.empty:
                print("âœ… RequÃªte exÃ©cutÃ©e avec succÃ¨s (aucun rÃ©sultat Ã  afficher).")
            else:
                print("\nğŸ“‹ RÃ©sultat :")
                print(tabulate(df, headers='keys', tablefmt='fancy_grid', showindex=False))

        except mysql.connector.Error as err:
            print(f"âŒ Erreur SQL : {err}")
        except Exception as e:
            print(f"âš ï¸ Erreur inattendue : {e}")
        finally:
            cursor.close()

# fonction principale
def main():
    print("ğŸ“˜ Bienvenue dans le programme de gestion de la base de donnÃ©es MariaDB")
    print("ğŸ› ï¸ Commandes SQL disponibles :")
    print("  - SELECT * FROM table_name;")
    print("  - INSERT INTO table_name (column1, column2) VALUES (value1, value2);")
    print("  - UPDATE table_name SET column1 = value1 WHERE condition;")
    print("  - DELETE FROM table_name WHERE condition;")
    print("  - STOP pour arrÃªter")


# fonction pour afficher la requete sous fome de graphique avec matplotlib

def plot_query_result(query_result):
    # Convertir le rÃ©sultat de la requÃªte en DataFrame
    df = pd.DataFrame(query_result)

    # CrÃ©er un graphique Ã  barres
    df.plot(kind='bar', x='column_name', y='value_column')
    plt.title('Titre du graphique')
    plt.xlabel('Nom de la colonne')
    plt.ylabel('Valeur de la colonne')
    plt.show()


# exÃ©cution du programme
if __name__ == "__main__":
    main()
    connect_to_database()
    plot_query_result()
```
